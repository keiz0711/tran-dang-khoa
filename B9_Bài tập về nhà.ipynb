Chương trình giải bài toán phương trình bậc 2

```python
import math

print("Chương trình giải phương trình bậc 2 dạng a*x^2 + b*x + c = 0")
a=int(input("Nhập hệ số a: "))
b=int(input("Nhập hệ số b: "))
c=int(input("Nhập hệ số c: "))

if a == 0:
    x0=-c/b
    print(f"Phương trình có 1 nghiệm x={x0}")
else:    
    d=b**2-4*a*c  
    if d<0:
        print(f"delta = {d}")
        print("Phương trình vô nghiệm")
    elif d==0:    
        print(f"delta = {d}")
        x1=-b/2*a
        print(f'Phương trình có nghiệm kép x={x1}')
    else:
        x11=(-b-math.sqrt(d))/2*a
        x12=(-b+math.sqrt(d))/2*a
        print(f"delta = {d}")
        print(f"""Phương trình có 2 nghiệm:   
        x1={'%.2f' % x11}
        x2={'%.2f' % x12}      
        """)




```
